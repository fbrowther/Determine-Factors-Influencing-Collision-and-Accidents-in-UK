#  Create new data with select columns

```python
new_data_acc = data_acc[['Accident_Index', 'Accident_Severity', 'Day_of_Week', 'Speed_limit', 'Road_Surface_Conditions', 'Weather_Conditions']].copy()
new_data_acc.head()
```

```python
new_data_veh = data_veh[['Accident_Index', 'Age_Band_of_Driver','Age_of_Vehicle','make','model','Sex_of_Driver', 'Vehicle_Manoeuvre']].copy() 
new_data_veh.head()
```

# Change the column names and drop NaN, data missing/ out of range values. 

```python
accident = new_data_acc.rename(
    columns = {"Accident_Index": "id",
                "Accident_Severity": "severity",
                "Speed_limit": "SpeedLimit",
                "Road_Surface_Conditions": "RoadConditions",
                "Day_of_Week": "day",
                "Weather_Conditions": "weather",
                
    }
)
```

```python
accident_data = accident.dropna()
accident_data
```

## Final accident_data file

```python
accident_data.set_index("id")
```

```python
vehicle0 = new_data_veh.rename(
    columns = {"Accident_Index": "id",
                "Age_Band_of_Driver": "driverAge",
                "Age_of_Vehicle": "vehicleAge",
                "make": "make",
                "model": "model",
                "Sex_of_Driver": "sex",
                "Vehicle_Manoeuvre": "VehicleManoeuvre"
    }
)
```

```python
vehicle0
```

```python
vehicle0.sex.unique()
```

```python
df_vehicle = vehicle0[vehicle0.sex != 'Data missing or out of range']
```

```python
df_vehicle1 = df_vehicle[df_vehicle.sex != 'Not known']
```
